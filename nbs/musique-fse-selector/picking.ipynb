```python
import json
from pathlib import Path

from bellem.utils import jprint
```

```python
data_dir = Path('../../data')
```

```python
auto_fs_examples_dir = data_dir / "raw/few-shot-examples/auto-interweave"
semi_fs_examples_dir = data_dir / "raw/few-shot-examples/semi"
```

```python
auto_fs_examples_mapping = dict()
for filepath in auto_fs_examples_dir.glob("*.json"):
    technique = filepath.stem
    with open(filepath, "r") as f:
        examples = json.load(f)
    auto_fs_examples_mapping[technique] = examples
```

```python
for record in auto_fs_examples_mapping['cot']:
    if record['n_hops'] == 4:
        answer = record['generation'].split('Answer:')[1]
        jprint(record)
        print("answer:", answer)
```

```python
picked_sample_ids = [
    "2hop__129019_11427",
    "3hop1__20673_720466_80004",
    "4hop1__632232_164954_6975_6833",
]
```

```python
def pick_samples(examples, sample_ids):
    example_mapping = {example['id']: example for example in examples}
    return [example_mapping[sample_id] for sample_id in sample_ids]
```

```python
for technique in auto_fs_examples_mapping:
    print(f"==== {technique} ====")
    picked_samples = pick_samples(auto_fs_examples_mapping[technique], picked_sample_ids)
    semi_filepath = semi_fs_examples_dir / f"{technique}.json"
    semi_examples = json.loads(semi_filepath.read_text())
    semi_examples.extend(picked_samples)
    semi_filepath.write_text(json.dumps(semi_examples, indent=2))
```
