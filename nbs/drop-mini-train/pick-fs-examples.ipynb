```python
import random
import json
import textwrap
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
from ipywidgets import HBox, widgets
```

```python
EXP_NAME = "butch-yips" # REASONING
EXP_NAME = "quack-conk" # CCOT
EXP_NAME = "cronk-doit" # DIRECT
```

```python
df = pd.read_json(f'results/results-{EXP_NAME}.jsonl', lines=True)
df.set_index('id', inplace=True, drop=False)
print(len(df))
df.head()
```

```python
success_mask = df['exact_match'] == 1
fail_df = df[~success_mask]
success_df = df[success_mask]
print("Fail:", len(fail_df))
print("Success:", len(success_df))
print("Total:", len(df))
print("Mean F1:", f"{df['f1'].mean():.3f}")
```

```python
def prepare_few_shot_example(row):
    assert row['exact_match'] == 1
    return {
        "id": row['id'],
        "context": row['hops'][0]['context'],
        "question": row['hops'][0]['question'],
        "generation": row['hops'][0]['qa_result']['generation'],
        "n_hops": len(row['hops']),
    },
```

```python
selected_ids = [
    "history_2374-be4944dd-6e6f-43c5-9e50-9e7761fc2f73",
    "history_4133-b6cc993b-2349-4d98-9b18-dcdd99649d62",
    "history_2149-c6201520-5504-427e-abfc-7f74d531bd53",
    "history_858-57fada68-04ec-4f31-8b57-cfac2807f34c",
    "history_1944-b8fbaceb-4b21-46ac-b22c-283c1301785e",
]

examples = [prepare_few_shot_example(df.loc[id]) for id in selected_ids]

with open(f'few-shot-examples-{EXP_NAME}.json', 'w') as f:
    json.dump(examples, f, indent=2)
```
